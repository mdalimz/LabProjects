```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
P=data_1[:, 0]
Volt=data_1[:, 1]
#plt.figure()
plt.plot(P,Volt,label='recorded data')
print(np.shape(Volt))
print(data_1[31,1])
print(data_1[[31],[1]])
```

```python
#imp
def sigmoid(P, L ,x0, k, b):
    y = L / (1 + np.exp(-k*(P-x0)))+b
    return (y)

p0 = [max(Volt)-min(Volt), np.median(P),1,min(Volt)] # this is an mandatory initial guess

popt, pcov = curve_fit(sigmoid, P, Volt,p0,bounds=(-np.inf,+np.inf), method='dogbox') #method=lm
#x = np.linspace(512, 412, 70)
y = sigmoid(P, *popt)
print(popt)
print(np.sqrt(pcov[1,1]))
print(pcov[1,1])
print(pcov)


plt.plot(P, Volt, 'o', label='data')
plt.plot(P,y, label='fit')
#plt.ylim(0, 1.05)
plt.legend(loc='best')
plt.title('Plot of Pressure versus Voltage of In')
plt.xlabel('Pressure(P mbar)')
plt.ylabel('voltage(V mV)')
plt.show()

'My age is' +str(42)
print('maxVolt=' +str (max(Volt)),'median='+str(np.median(P)),'1','minVolt='+str(min(Volt)))


```
