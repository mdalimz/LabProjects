```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit 
from scipy.signal import find_peaks
```

```python
data_4=np.loadtxt("G7//New folder//cp0,1.dat", comments="#", unpack=False, skiprows=1)
```

```python
energy=data_4[:, 0]
count=data_4[:, 1]
plt.figure()
plt.plot(energy,count,label='recorded data')
```

```python
peaks, _ = find_peaks(count) 
plt.scatter(energy[peaks], count[peaks])
def f(x,a,b,x0,y0): 
    return (0.752* (np.sign(energy[peaks]-x0) + 1)*a*np.exp(-b*(energy[peaks]-x0))) +y0
pars, cov = curve_fit(f, xdata=energy[peaks], ydata=count[peaks], p0=[5,3,0.05,0.0001],bounds=(-np.inf, np.inf),maxfev=10000)#, bounds=(-np.inf, np.inf))
plt.plot(energy[peaks], f(energy[peaks], *pars), 'k--')
#print(dtype(peaks))
```

```python
energy=data_4[:, 0]
count=data_4[:, 1]
def f(x,a,b,x0,y0,t): 
    return (0.752* (np.sign(x-x0) + 1)*a*np.cos(x)*np.exp(-b*(x-x0))) +y0
```

```python
#fitting function
def f(x,a,b,x0,y0,t): 
    return ((a/2)* (np.sign(x-x0) + 1)*np.exp(-b*(x-x0))*(np.cos(t*(x-x0)))**2) +y0
pars, cov = curve_fit(f, xdata=energy, ydata=count, p0=[0.5, 0.5,0.05,0.01,0.01],bounds=(-np.inf, np.inf),maxfev=10000)#bounds=(-np.inf, np.inf)          
plt.plot(energy, f(energy, *pars), 'k--',label= 'Fitted curve') 
plt.plot(energy,count,'g',label='Carr-Purcell of CuSo4:0.1M')
plt.legend(loc='best')
plt.title('Carr-Purcell of CuSo4:0.1M')
plt.xlabel('time(s)')
plt.ylabel('Amplitude')
plt.show()
print(cov)
print(pars)
print(np.sqrt(cov[1,1]))

```
