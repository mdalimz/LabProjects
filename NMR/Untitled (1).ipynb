## Fourier Transform

```python
#fft
import numpy as np
import matplotlib.pyplot as plt
#import scipy
from scipy.fft import fft, ifft
import cmath
```

```python
#https://www.youtube.com/watch?v=R_4GuyJTzMo&t=2201s
n = 30
t = np.arange(0,2*np.pi,2*np.pi/n)
f = np.sin(t)
F = fft(f)
plt.subplot(2,1,1)
plt.scatter(t,f,marker='o')
plt.subplot(2,1,2)
plt.scatter(range(0,len(F)),F.real, marker='o',color='green')
# plt.subplot(2,1,3)
plt.scatter(range(0,len(F)),F.imag, marker='o',color='purple')
# plt.subplot(2,1,4)
plt.scatter(range(0,len(F)),np.abs(F), marker='o',color='blue')
plt.show()
print(len(F))
```

```python
Finv = ifft(F)
plt.subplot(2,1,2)
plt.scatter(range(0,len(Finv)),Finv.real, marker='o',color='green')
plt.scatter(range(0,len(Finv)),Finv.imag, marker='o',color='purple')
plt.scatter(range(0,len(Finv)),np.abs(Finv), marker='o',color='blue')
plt.show()
```

```python
#fs is sampling frequency
fs = 100.0
time = np.linspace(0,10,int(10*fs),endpoint=False)
# print(1/(time[1]-time[0]))
print(time[200])

#wave is the sum of sine wave(1Hz) and cosine wave(10 Hz)
wave = np.sin(np.pi*time)+ np.cos(np.pi*time)
#print(wave)
#wave = np.exp(2j * np.pi * time )

plt.plot(time, wave)
plt.xlim(0,10)
plt.xlabel("time (second)")
plt.title('Original Signal in Time Domain')

plt.show()
print(len(wave))
i = np.where(time==2)
print(i)
w = np.where(wave==1.00000000e+00)
print(w)
```

```python
# Compute the one-dimensional discrete Fourier Transform.

fft_wave = np.fft.fft(wave)

# Compute the Discrete Fourier Transform sample frequencies.

fft_fre = np.fft.fftfreq(n=wave.size, d=1/fs)

plt.subplot(211)
plt.plot(fft_fre, fft_wave.real, label="Real part")
plt.xlim(-50,50)
plt.ylim(-600,600)
plt.legend(loc=1)
plt.title("FFT in Frequency Domain")

plt.subplot(212)
plt.plot(fft_fre, fft_wave.imag,label="Imaginary part")
plt.legend(loc=1)
plt.xlim(-50,50)
plt.ylim(-600,600)
plt.xlabel("frequency (Hz)")
# plt.subplot()
# plt.plot(fft_fre, np.abs(fft_wave.imag),label="Imaginary part")
# print(fft_wave)
plt.show()
print(wave.size)
print(len(fft_fre))
```

```python
#test_code_2_Zehn.py
import numpy as np
import matplotlib.pyplot as plt

signal = np.array([-2, 8, 6, 4, 1, 0, 3, 5], dtype=float)
fourier = np.fft.fft(signal)
n = signal.size
timestep = 0.1
freq = np.fft.fftfreq(n, d=timestep)

plt.figure(0)
plt.plot(range(n),signal) #'-o', for black colour '-0','-.'
plt.show()

plt.figure(1)
plt.scatter(freq,fourier.real)
plt.show()

plt.figure(2)
plt.scatter(freq,fourier.imag)
plt.show()
```
