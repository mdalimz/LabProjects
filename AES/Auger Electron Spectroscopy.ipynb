# Uncleaned Surface

```python
#Loading dataset
data1=np.loadtxt("D:/LAB/Auger/G7_data/unclean_surface 2", unpack=False, skiprows=8)
```

- unpack=False:
This parameter indicates that the data should be returned as a 2D array (default behavior). If unpack were set to True, the columns of the data would be returned as separate arrays.

- skiprows=8:
This parameter specifies that the first 8 lines of the file should be skipped before reading the data. This is useful if the file has a header or some initial metadata that you want to ignore.

```python
from scipy.ndimage import gaussian_filter1d
energy=data1[:, 0]  #extracts all values from the first column of data 1
count=data1[:, 1]   #extracts all values from the second column...........

#The gaussian_filter1d function smooths the data by averaging each point with its neighbors, weighted by a Gaussian function.A larger sigma value results in more smoothing.
y3 = gaussian_filter1d(count, 3)
y6 = gaussian_filter1d(count, 13)

##plt.plot(energy,count,label='recorded data',linestyle='dashed')
plt.figure(1, figsize=(8,4))
plt.plot(energy,count,label='recorded data')
plt.plot(energy,y3, '--', label='filtered, sigma=3') #The sigma parameter determines the standard deviation of the Gaussian kernel, which affects the degree of smoothing. A higher sigma results in more smoothing.
plt.plot(energy,y6, ':', label='filtered, sigma=13')
plt.legend()
plt.xlabel('energy(eV)')
plt.ylabel('dN/dE)')
plt.show()
```

```python
from scipy.signal import savgol_filter
yhat = savgol_filter(count, 51, 11) # window siz
plt.figure(2,figsize=(8,5))
plt.plot(energy,count,label='recorded data')
plt.plot(energy,yhat, color='red',label='smoothed data')
plt.plot([72 for x in energy],count,'g',label='Ba',)
plt.plot([516 for x in energy],count,'b',label='O',)
plt.plot([1231.4 for x in energy],count,'y',label='As',)
plt.plot([1393.5 for x in energy],count,'m',label='Al',)
plt.plot([266.4 for x in energy],count,'c',label='C',)
plt.legend(loc='best')
plt.title('Spectrum of Dirty Sample')
plt.xlabel('energy(eV)')
plt.ylabel('dN/dE)')

plt.show()
```

```python
#https://www.youtube.com/watch?v=Iy_HXyMaLq4
from scipy.misc import derivative
# def fun(u,a):
#     return -2*a*x*exp(-a*x**2)

def fun(u,a=2):
     return np.exp(-a*u**2)


def deri(u):
    return derivative(fun,u)

x=np.linspace(-2,2,50)
plt.figure()
plt.plot(x,deri(x),label = 'derivative fun')
plt.plot(x,fun(x,2),label = 'fun')
plt.axhline(y = 0,color = 'r', linestyle = '-')
plt.axvline(x = 0,color = 'g', linestyle = '-')

plt.legend()
plt.show()
```

```python
#https://stackoverflow.com/questions/31556446/how-to-draw-axis-in-the-middle-of-the-figure
import math
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a


x = np.arange(-10., 10., 0.2)
sig = sigmoid(x)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(x,sig)
plt.show()
```
